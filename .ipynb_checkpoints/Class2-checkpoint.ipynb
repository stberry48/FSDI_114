````python
# simplified implementation of stack (relying on built-ins)

class Stack:
    def __init__(self):
        self.items = []
    
    def push(self,element):
        self.items.append(element)
    
    def pop(self):
        return self.items.pop()
    
    # nice to have methods
    def peek(self):
        return self.items[len(self.items)-1]
    
    def size(self):
        return len(self.items)
    
    def is_empty(self):
        return self.items ==[]

# shift+enter when done typing
````

````python
class QueueII:
    def __init__(self):         
        self.queue = []          # An empty list to store the queue elements
        self.front = 0           # An index pointing to the front of the queue where items are dequeued
        self.rear = 0            # An index pointing to the rear of the queue where new items are added. When items are added to the list it increases the rear(see line 9).

    def enqueue(self, item):
        self.queue.append(item)  # Adds new item to end of list/queue
        self.rear += 1

    def dequeue(self):
        if self.is_empty():      # Checks if queue is empty before attempting to dequeue(front)
            return None
        item = self.queue[self.front]  # If not empty retrieves the item from the front index
        self.front += 1               # Increments the front index to point to next item (think: next in line at the counter)
        return item                   # Returns the dequeued(front)items (does this mean it's returning to the back or front)?

    def is_empty(self):
        return self.front == self.rear  # queue is empty when front and rear are equal

    def size(self):
        return self.rear - self.front    # Calculates the number of items in the queue by subtracting front from rear

    def peek(self):
        if self.is_empty():
            return None
        return self.queue[self.front]     # Returns item at the front of the queue without removing it.

    def display(self):
        if self.is_empty():
            print("Queue is empty")
        else:
            print("Queue:", self.queue[self.front:self.rear]) # Prints the queue contents from front to rear

# Testing the QueueII class
q = QueueII()
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
q.display()  # Output: Queue: [1, 2, 3]

print(q.dequeue())  # Output: 1
q.display()  # Output: Queue: [2, 3]

print(q.peek())  # Output: 2
print(q.size())  # Output: 2
print(q.is_empty())  # Output: False

q.dequeue()
q.dequeue()
print(q.is_empty())  # Output: True
q.display()  # Output: Queue is empty

            
````
